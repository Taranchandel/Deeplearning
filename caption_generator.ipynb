```python
import whisper

model = whisper.load_model("base")
result = model.transcribe("WhatsApp Audio 2026-02-24 at 7.36.53 PM.aac")

print(result["text"])
```

```python
for segment in result["segments"]:
    start = segment["start"]
    end = segment["end"]
    text = segment["text"]

    print(f"[{start:.2f}s - {end:.2f}s] {text}")
```

```python
raw_text = """
[0.00s - 5.32s]  Hey it's a Taran, I think that video could be very useful for us.
[5.32s - 12.08s]  So I sharing it, that is a Stanford video, there is a 20 minute conversation among students.
[12.08s - 17.96s]  They are discussing something more interesting like how they are using charge-gpt and so more.
[17.96s - 20.16s]  I think we should watch this video.
[20.16s - 22.80s]  I think we must watch this video.
[22.80s - 29.32s]  If you are comfortable in Hindi, then you can also change your audio track in Hindi from YouTube settings.
[29.32s - 31.32s]  Yeah, I think that's great.
[31.32s - 32.32s]  Thank you.
"""
import re

def format_time(seconds):
    seconds = float(seconds)
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    millis = int((seconds - int(seconds)) * 1000)
    return f"{hours:02}:{minutes:02}:{secs:02},{millis:03}"

lines = raw_text.strip().split("\n")

with open("converted_captions.srt", "w", encoding="utf-8") as f:
    counter = 1
    for line in lines:
        match = re.match(r"\[(.*?)s - (.*?)s\]\s*(.*)", line)
        if match:
            start, end, text = match.groups()
            f.write(f"{counter}\n")
            f.write(f"{format_time(start)} --> {format_time(end)}\n")
            f.write(f"{text.strip()}\n\n")
            counter += 1

print("SRT file created successfully!")
```
